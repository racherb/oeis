```
import math
import random
import nevergrad as ng

# Bendford Probability
def p_benford(d):
    p = math.log10(1+(1/d)) 
    return p

# Numerical Serie Builder
def succession_builder(pattern):
    lsup = len(pattern)
    serie = []
    for b in range (1, lsup+1):
        for j in range(0, lsup):
            suc = pattern[j:j+b]
            if len(suc)==b:
                serie.append(suc)
    return serie

# Gain of Representativeness
def gain_rep(serie, linf, lsup):
    gain = 0
    for num in range(linf, lsup):
        if serie.find(str(num)) >= 0:
            gain = gain + 1
    return gain

# Pattern Builder
def pattern_builder(size, linf, lsup):
    rnd_lst = []
    for i in range(1, size):
        n = random.randint(linf, lsup)
        rnd_lst.append(str(n))
    return rnd_lst

def make_succesion(oeis):
    serie = oeis.replace(',', '')
    serie_lst = succession_builder(serie)
    return ''.join(serie_lst)

def attrs_rep(serie, linf, lsup):
    size = len(serie)
    gain = gain_rep(serie, linf, lsup)
    ratio = gain/size

    serie_ext = serie + serie[::-1]
    size_ext = len(serie_ext)
    gain_ext = gain_rep(serie_ext, linf, lsup)
    ratio_ext = gain_ext/size_ext

    return size, gain, ratio, size_ext, gain_ext, ratio_ext

```
